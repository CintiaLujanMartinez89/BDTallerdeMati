# [dbo].[ObtenerMotosPorDni]

```sql
/****** Object:  StoredProcedure [dbo].[ObtenerMotosPorDni]    Script Date: 1/10/2024 09:45:48 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[ObtenerMotosPorDni]
    @dni VARCHAR(10)
AS
BEGIN
    SELECT 
        Patente_Moto_CXM,
        Marca_M,
        Modelo_M,
      Nombre_S,
   FORMAT(Fecha_Realizacion_HXS, 'dd/MM/yyyy') AS Fecha_Realizacion_HXS
    FROM CLIENTES_X_MOTOS AS cxm
   INNER JOIN Motos AS m ON cxm.Patente_Moto_CXM = m.Patente_Moto_M
   LEFT JOIN HISTORIAL_X_SERVICIOS hxs ON m.Patente_Moto_M = hxs.Patente_Moto_HXS
   LEFT JOIN SERVICIOS AS s ON s.Id_Servicio_S = hxs.Id_Servicio_HXS
    WHERE Dni_CXM =@dni AND Estado_M=1;
END;
GO
```

# [dbo].[ObtenerTablaHistorial]

```sql
/****** Object:  StoredProcedure [dbo].[ObtenerTablaHistorial]    Script Date: 1/10/2024 09:45:48 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[ObtenerTablaHistorial]

AS
BEGIN
 
SELECT Dni_HXS,Nombre_CLI+' '+Apellido_CLI as Nombre_CLI,Detalle_HXS,Patente_Moto_HXS,Marca_M,Modelo_M, FORMAT(Fecha_Realizacion_HXS, 'dd/MM/yyyy') AS Fecha_Realizacion_HXS,Kilometraje_HXS,Nombre_S
FROM HISTORIAL_X_SERVICIOS inner join CLIENTES on Dni_HXS=Dni_CLI
							inner join MOTOS on Patente_Moto_HXS=Patente_Moto_M
							inner join SERVICIOS on Id_Servicio_S=Id_Servicio_HXS
END
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[spAgregarHistorial]    Script Date: 1/10/2024 09:45:48 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[spAgregarHistorial]
    @Dni_H VARCHAR(10),
    @Patente_Moto_H VARCHAR(10),
    @Estado_H BIT
AS
BEGIN
    DECLARE @NuevoId VARCHAR(10);
    
    -- Obtener el último Id_Historial_H
    DECLARE @UltimoId INT;
    SELECT @UltimoId = MAX(CAST(SUBSTRING(Id_Historial_H, 2, LEN(Id_Historial_H) - 1) AS INT))
    FROM HISTORIAL;

    -- Generar el nuevo Id_Historial_H
    SET @NuevoId = 'H' + CAST(@UltimoId + 1 AS VARCHAR);

    -- Insertar el nuevo registro
    INSERT INTO [dbo].[HISTORIAL] ([Id_Historial_H], [Dni_H], [Patente_Moto_H], [Estado_H])
    VALUES (@NuevoId, @Dni_H, @Patente_Moto_H, @Estado_H);
END
GO
```

# [dbo].[spCambiarAsistencia]

```sql
/****** Object:  StoredProcedure [dbo].[spCambiarAsistencia]    Script Date: 1/10/2024 09:45:48 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[spCambiarAsistencia]
    @IdTurno INT,
    @Dia DATE,
    @Hora TIME
AS
BEGIN
    SET NOCOUNT ON;

    UPDATE TURNOS
    SET Asistencia_T = CASE 
                          WHEN Asistencia_T = 1 THEN 0 
                          ELSE 1 
                       END
    WHERE Id_Turno_T = @IdTurno AND Dia_T = @Dia AND Hora_T = @Hora;
	SELECT @@ROWCOUNT AS FilasAfectadas;
END
GO
```

# [dbo].[spInsertarCliente]

```sql
/****** Object:  StoredProcedure [dbo].[spInsertarCliente]    Script Date: 1/10/2024 09:45:48 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[spInsertarCliente]
    @Dni VARCHAR(10),
    @Nombre VARCHAR(20),
    @Apellido VARCHAR(20),
    @Domicilio VARCHAR(50),
    @Telefono VARCHAR(15),
    @Email VARCHAR(30),
    @HashContraseña VARCHAR(255)
AS
BEGIN
    -- Verificar si el DNI ya existe
    IF EXISTS (SELECT 1 FROM CLIENTES WHERE Dni_CLI = @Dni)
    BEGIN
        -- El DNI ya existe, devolver 2
        RETURN 2;
    END
    ELSE
    BEGIN
        -- Insertar el nuevo cliente
        INSERT INTO CLIENTES (Dni_CLI, Nombre_CLI, Apellido_CLI, Domicilio_CLI, Telefono_CLI, Email, Estado_CLI, HashContraseña_CLI)
        VALUES (@Dni, @Nombre, @Apellido, @Domicilio, @Telefono, @Email, 1, @HashContraseña);
        
        -- Devolver 1 para indicar que la inserción fue exitosa
        RETURN 1;
    END
END
GO
```

# [dbo].[spObtenerTurnosAsignados]

```sql
/****** Object:  StoredProcedure [dbo].[spObtenerTurnosAsignados]    Script Date: 1/10/2024 09:45:48 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[spObtenerTurnosAsignados]
AS
BEGIN
 

    SELECT 
        Id_Turno_T,
       FORMAT(Dia_T, 'dd/MM/yyyy') AS Dia_T,
          Hora_T, 
        Nombre_CLI + ' ' + Apellido_CLI AS Nombre_CLI,
        Patente_Moto_T,
        Marca_M,
        Modelo_M,
        Nombre_S,
        Observacion_T,
        Asistencia_T
    FROM 
        TURNOS
    INNER JOIN 
        CLIENTES ON Dni_CLI = Dni_T
    INNER JOIN 
        CLIENTES_X_MOTOS ON Dni_CXM = Dni_T
    INNER JOIN 
        MOTOS ON Patente_Moto_M = Patente_Moto_T
    INNER JOIN 
        SERVICIOS ON Id_Servicio_S = Id_Servicio_T
    GROUP BY 
        Id_Turno_T,
        Dia_T,
        Hora_T,
        Nombre_CLI,
        Apellido_CLI,
        Marca_M,
        Modelo_M,
        Observacion_T,
        Nombre_S,
        Asistencia_T,
        Patente_Moto_T
END
GO
```
